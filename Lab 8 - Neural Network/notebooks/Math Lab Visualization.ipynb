```python
import numpy as np
import pandas as pd
import plotly.express as px

# 1. Define the number of samples (balanced dataset)
m = 1000
samples_per_class = m // 2

# 2. Generate samples for Class 0: X | Y=0 ~ N(2, 2)
# We interpret this as Mean = [2, 2] and Variance = 2
mean0 = [2, 2]
cov0 = [[2, 0], [0, 2]]  # Isotropic covariance
class0_data = np.random.multivariate_normal(mean0, cov0, samples_per_class)
df0 = pd.DataFrame(class0_data, columns=['Feature 1', 'Feature 2'])
df0['Label'] = 'Class 0'

# 3. Generate samples for Class 1: X | Y=1 ~ N(0, 3)
# We interpret this as Mean = [0, 0] and Variance = 3
mean1 = [0, 0]
cov1 = [[3, 0], [0, 3]]
class1_data = np.random.multivariate_normal(mean1, cov1, samples_per_class)
df1 = pd.DataFrame(class1_data, columns=['Feature 1', 'Feature 2'])
df1['Label'] = 'Class 1'

# 4. Combine into a single DataFrame
df = pd.concat([df0, df1])

# 5. Visualize with Plotly Express
fig = px.scatter(
    df,
    x='Feature 1',
    y='Feature 2',
    color='Label',
    title='Synthetic Dataset: $X | Y=0 \sim \mathcal{N}(2,2)$ vs $X | Y=1 \sim \mathcal{N}(0,3)$',
    labels={'Label': 'Class'},
    template='plotly_white',
    opacity=0.7
)

fig.show()
```
