## Question 2: Building Text Processing Functions

### Question 2.1: Building a Data Cleaning Function

**Description:** The function removes digits and keeps only the characters "A-Za-z(),!?\'\`".


```python
# clean text
def clean_text(text):
    #### YOUR CODE HERE ####
    text = re.sub(r"[^A-Za-z(),!?\'\` ]", "", text)
    return text
    #### END YOUR CODE #####
```

```python
test_text = "123! Let's go! @@"
print(clean_text(test_text))
```

### Question 2.2: Function to Convert Text to Lowercase

```python
# make all text lowercase
#### YOUR CODE HERE ####
def text_lowercase(text):
    text = text.lower()
    return text
#### END YOUR CODE #####
```

```python
test_text = "REALLL? I will check that soon."
print(text_lowercase(test_text))
```

### Question 2.3: Building a Function to Remove Punctuation

```python
# remove punctuation
def remove_punctuation(text):
    #### YOUR CODE HERE ####
    return "".join(char for char in text if char not in string.punctuation)
    #### END YOUR CODE #####
```

```python
test_text = "REALLL? I will check that soon."
print(remove_punctuation(test_text))
```

### Question 2.4: Tokenization

```python
# tokenize
nltk.download('punkt_tab')
def tokenize(text):
    #### YOUR CODE HERE ####
    tokens = word_tokenize(text)
    return tokens
    #### END YOUR CODE #####
```

```python
test_text = "REALLL? I will check that soon."
print(tokenize(test_text))
```

### Question 2.5: Removing Stopwords







```python
# remove stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
#### YOUR CODE HERE ####
def remove_stopwords(text):
    tokens = tokenize(text)
    tokens_no_stopwords = [token for token in tokens if token.lower() not in stop_words]
    return " ".join(tokens_no_stopwords)
#### END YOUR CODE #####
```

```python
test_text = "The dog is laughing"
print(remove_stopwords(test_text))
```

### Question 2.6: Building a Lemmatization Function







```python
#### YOUR CODE HERE ####

# lemmatize

# Add library
from nltk.corpus import wordnet
from nltk import pos_tag
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')

# Function to convert NLTK POS tags to WordNet POS tags
def wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        # Default to noun if no clear mapping is found
        return wordnet.NOUN


lemmatizer = WordNetLemmatizer()
def lemmatize(text):
    tokens = tokenize(text)
    tagged_tokens = pos_tag(tokens)

    lemmatized_sentence = []
    for token, tag in tagged_tokens:
      wn_tag = wordnet_pos(tag)
      lemma = lemmatizer.lemmatize(token, wn_tag)
      lemmatized_sentence.append(lemma)
    return lemmatized_sentence


#### END YOUR CODE #####
```

```python
test_text = "I have completed some questions, so i will move on harder problems."
print(lemmatize(test_text))
```

### Question 2.7: Building a Preprocessing Function
**Hint:** This function will call the functions written above.

```python
def preprocessing(text):
#### YOUR CODE HERE ####
    text = clean_text(text)
    text = text_lowercase(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text)
    text = lemmatize(text)
    return text
#### END YOUR CODE #####
```

## Question 3: Preprocessing for Input Text
**Overview:** Using the functions above, preprocess the initial text.

```python
#### YOUR CODE HERE ####
preprocessed_docs = []
for doc_content in docs_content:
  preprocessed_docs.append(preprocessing(doc_content))

for idx in range(0, len(preprocessed_docs), 600):
  print(docs_content[idx])
  print(preprocessed_docs[idx], "\n")
```
