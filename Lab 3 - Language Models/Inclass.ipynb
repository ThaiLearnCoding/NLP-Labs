# Language model based on n-grams using the Laplace smoothing

```python
from collections import Counter
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt') # Download punkt for sentence tokenization
nltk.download('punkt_tab')
```

```python
def build_LMs(string):

  sentences = sent_tokenize(string)
  LM_1gram = Counter()
  LM_2gram = Counter()
  LM_3gram = Counter()

  # Process each sentence
  for sentence in sentences:
    words = word_tokenize(sentence.lower())
    words = [word for word in words if word.strip()]

    # Build 1-gram LM
    LM_1gram.update(words)

    # Build 2-gram LM
    words_for_2gram = ['<s>'] + words + ['</s>']
    bigrams = list(ngrams(words_for_2gram, n = 2))
    LM_2gram.update(bigrams)

    # Build 3-gram LM
    words_for_3gram = ['<s>', '<s>'] + words + ['</s>']
    trigrams = list(ngrams(words_for_3gram, n = 3))
    LM_3gram.update(trigrams)

  return LM_1gram, LM_2gram, LM_3gram
```

```python
LM_1gram, LM_2gram, LM_3gram = build_LMs(txt_content)
```

## 1-gram Language Model

```python
def calculate_P_1gram(LM_1gram, word_i):
  P = (LM_1gram[word_i] + 1) / (len(LM_1gram) + LM_1gram.total())
  return P
```

```python
def test_string_1gram(test_word):
  test_word = test_word.lower()
  print(f"Number of occurence of the string \"{test_word}\" in the corpus: {LM_1gram[test_word]}")
  P_test_word = calculate_P_1gram(LM_1gram, test_word)
  print(f"Smoothed probability of the word \"{test_word}\" using Laplace smoothing: {P_test_word:.4e}")
```

```python
# Out-of-distribution word
test_string_1gram("bruh")

# Common word
test_string_1gram("I")
```

## 2-gram Language Model

```python
def calculate_P_2gram(LM_1gram, LM_2gram, word_i1, word_i):
  P = (LM_2gram[(word_i1, word_i)] + 1) / (LM_1gram[word_i1] + len(LM_1gram))
  return P
```

```python
def test_string_2gram(test_string):
  test_word_i1, test_word_i = word_tokenize(test_string)

  test_word_i1= test_word_i1.lower()
  test_word_i= test_word_i.lower()

  print(f"Number of occurence of the string \"{test_string}\" in the corpus: {LM_2gram[(test_word_i1, test_word_i)]}")
  P_test_word = calculate_P_2gram(LM_1gram, LM_2gram, test_word_i1, test_word_i)
  print(f"Smoothed probability of the word \"{test_string}\" using Laplace smoothing: {P_test_word:.4e}")
```

```python
# Out-of-distribution
test_string_2gram("I her")

# Common
test_string_2gram("I'm")
```

## 3-gram Language Model

```python
def calculate_P_3gram(vocab_size, LM_2gram, LM_3gram, word_i2, word_i1, word_i):
  P = (LM_3gram[(word_i2, word_i1, word_i)] + 1) / (LM_2gram[(word_i2, word_i1)] + vocab_size)
  return P
```

```python
def test_string_3gram(test_string):
  test_word_i2, test_word_i1, test_word_i = word_tokenize(test_string)

  test_word_i2= test_word_i2.lower()
  test_word_i1= test_word_i1.lower()
  test_word_i= test_word_i.lower()

  print(f"Number of occurence of the string \"{test_string}\" in the corpus: {LM_3gram[(test_word_i2, test_word_i1, test_word_i)]}")
  P_test_word = calculate_P_3gram(len(LM_1gram), LM_2gram, LM_3gram, test_word_i2, test_word_i1, test_word_i)
  print(f"Smoothed probability of the word \"{test_string}\" using Laplace smoothing: {P_test_word:.4e}")
```

```python
# Out-of-distribution
test_string_3gram("I'm handsome")

# Common
test_string_3gram("I'm happy")
```

# Calculating probability of the sentence and the perplexity

## Probability of the sentence

```python
def process_sentence(sentence, type_n):
  tokens = word_tokenize(sentence.lower())
  tokens = [token for token in tokens if token.strip()]

  if type_n == 2:
    tokens = ['<s>'] + tokens + ['</s>']
  elif type_n == 3:
    tokens = ['<s>', '<s>'] + tokens + ['</s>']

  return tokens
```

```python
def sentence_prob(sentence, type_n, LM_1gram = None, LM_2gram = None, LM_3gram = None):
  '''
  If type_n = n, please fill in LM_igram with i = 1, 2, ..., n-1
  '''

  # Unsupported
  if type_n not in [1, 2, 3]:
    print(f"Do not support this type of language model: {type_n}-gram ! Retry :)")
    return 0.0

  # Process the sentence
  tokens = process_sentence(sentence, type_n)
  n_gram_list = list(ngrams(tokens, n = type_n))

  # Calculate the prob
  P = 1.0

  if type_n == 1:
    for n_gram in n_gram_list:
      P *= calculate_P_1gram(LM_1gram, n_gram[0])
  elif type_n == 2:
    for n_gram in n_gram_list:
      P *= calculate_P_2gram(LM_1gram, LM_2gram, n_gram[0], n_gram[1])
  else:
    for n_gram in n_gram_list:
      P *= calculate_P_3gram(len(LM_1gram), LM_2gram, LM_3gram, n_gram[0], n_gram[1], n_gram[2])
  return P
```

```python
def test_sentence_prob(sentence, LM_1gram, LM_2gram, LM_3gram):
  print(f"Probability of the sentence \"{sentence}\" using")

  P_1 = sentence_prob(sentence, 1, LM_1gram)
  print(f"{1}-gram LM: {P_1:.4e}")

  P_2 = sentence_prob(sentence, 2, LM_1gram, LM_2gram)
  print(f"{2}-gram LM: {P_2:.4e}")

  P_3 = sentence_prob(sentence, 3, LM_1gram, LM_2gram, LM_3gram)
  print(f"{3}-gram LM: {P_3:.4e}\n")

  return P_1, P_2, P_3
```

## Perplexity of the sentence

```python
import math

def sentence_perplexity(sentence, sentence_prob, type_n):
  tokens = process_sentence(sentence, type_n)
  perplexity = pow(sentence_prob, 1/len(tokens))
  return perplexity
```

```python
def test_perplexity(sentence, P_1, P_2, P_3):
  print(f"Perplexity of the sentence \"{sentence}\" using")

  PP_1 = sentence_perplexity(sentence, P_1, 1)
  print(f"{1}-gram LM: {PP_1:.4e}")

  PP_2 = sentence_perplexity(sentence, P_2, 2)
  print(f"{2}-gram LM: {PP_2:.4e}")

  PP_3 = sentence_perplexity(sentence, P_3, 3)
  print(f"{3}-gram LM: {PP_3:.4e}\n")

  return PP_1, PP_2, PP_3
```

## Suggestions for Further Refinement and Improvement for the In-class code

Here are some ideas to enhance your language model and move towards the spelling error correction objective:

1.  **More Sophisticated Smoothing Techniques:** While Laplace smoothing is a good starting point, for more realistic language modeling, consider exploring:
    *   **Kneser-Ney Smoothing:** Often considered the best smoothing technique for N-gram models, especially for large vocabularies.
    *   **Witten-Bell Smoothing** or **Good-Turing Discounting:** Other alternatives that often perform better than simple add-k smoothing.

2.  **Handling Out-Of-Vocabulary (OOV) Words:** Laplace smoothing assigns a non-zero probability to OOV words, but for practical applications, you might want more explicit OOV handling:
    *   **`<unk>` Token:** Replace rare words (e.g., words occurring less than N times) in your training data with a special `<unk>` token. This allows the model to learn probabilities for unknown words.

3.  **Tokenization Improvements:**
    *   **Contractions:** `nltk.word_tokenize` often handles contractions well (e.g., "I'm" -> "I", "'m"), but you might want to standardize them further or keep them as single tokens depending on your specific needs.
    *   **Punctuation Handling:** You've made a choice to keep punctuation. Depending on the task, you might want to remove all punctuation or treat it differently (e.g., normalizing multiple spaces, standardizing quotation marks).

4.  **Efficiency for Larger Datasets:** For very large text corpora:
    *   **Memory Usage:** Storing all n-grams in `Counter` objects can become memory-intensive. Consider using techniques like memory-mapped files or more efficient data structures if you encounter memory limits.

5.  **Perplexity Interpretation:** Perplexity is a good intrinsic evaluation metric. Lower perplexity generally indicates a better model. When comparing models, a significant difference in perplexity values is usually more meaningful than very small differences.

6.  **Developing Spelling Correction:** To implement the actual spelling error correction:
    *   **Candidate Generation:** For a given misspelled word, you'll need a way to generate potential correct words (e.g., using edit distance, keyboard proximity, phonetic similarity).
    *   **Candidate Ranking:** Use your language model to rank these candidates. A common approach is to choose the candidate `c` that maximizes `P(c | context)`. This means calculating the probability of the entire sentence with the candidate word in place, and selecting the candidate that yields the highest sentence probability (or lowest perplexity).
    *   **Error Models:** More advanced spell checkers often include an "error model" `P(misspell | correct_word)` which estimates the probability of typing a specific misspelling given a correct word. The final choice would then maximize `P(correct_word | context) * P(misspell | correct_word)`.

7.  **Consider Libraries:** For complex NLP tasks, leveraging existing libraries like `spaCy` or `Hugging Face Transformers` can provide highly optimized and pre-trained models for tasks beyond simple n-grams, including more robust tokenization and contextual embeddings that are superior for tasks like spelling correction.
